```python
earlier = pd.read_csv('1819.csv')
later = pd.read_csv('1920.csv')

skater = pd.read_csv('1920skater.csv')
```

```python


# write code to map the 'player' collumn in the sjater database from the current string to a different string which i will define
def change_name(given_name):
    given_name = given_name.replace("'", "")
    elements = given_name.split(' ')
    return elements[1] + ', ' + elements[0]

skater['Player'] = skater['Player'].apply(change_name)
```

```python
skater
```

```python


# Remove duplicates for entries with the team 'TOT'
skater.drop_duplicates(subset='Player', keep='first')

```

```python
skater.filter(items=['Player', 'FOW', 'FOL'])
```

```python

later = later.filter(items=['Name', 'FO', 'C', 'LW', 'RW', 'LD', 'RD'])
earlier = earlier.filter(items=['Name', 'FO', 'C', 'LW', 'RW', 'LD', 'RD'])


```

```python
def replace(string):
    return string.replace("'", "")
later['Name'] = later['Name'].apply(replace)
earlier['Name'] = earlier['Name'].apply(replace)
```

```python
# here we will prepare the x and y data for the model
x_data = []
y_data = []

skaters_names = later['Name'].tolist()
data_names = skater['Player'].tolist()
skaters_names = set(data_names) | set(skaters_names)




prev_data_names = earlier['Name'].tolist()

for name in skaters_names:
    print(name)
    prev = 2
    if name in prev_data_names:
        prev = earlier.query(f"Name == '{name}'").iloc[0]['FO']
    ind_skater_data = skater.query(f"Player == '{name}'")
    later_query = later.query(f"Name == '{name}'")

    if ind_skater_data.empty or later_query.empty:
        print("no data for '{name}'".format(name=name))
        continue

    wins = ind_skater_data.iloc[0]['FOW']
    losses = ind_skater_data.iloc[0]['FOL']
    actual_rating = later_query.iloc[0]['FO']

    print(actual_rating)


    x_data.append((prev, wins, losses))
    y_data.append(actual_rating)

print(y_data)
x_data = np.array(x_data)
y_data = np.array(y_data)
```

```python

model = models.Sequential([
    layers.Dense(3, activation='tanh', input_shape=(3,), kernel_initializer='normal'),
    layers.Dense(3, activation='tanh'),
    layers.Dense(3, activation='tanh'),
    layers.Dense(3, activation='tanh'),
    layers.Dense(1, activation='linear')
])
```

```python
model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['mean_squared_error'])
```

```python
model.summary()
```

```python
history = model.fit(x_data, y_data, epochs=20, validation_split=.2)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train loss', 'test loss'], loc='center right')
plt.show()
```

```python


# Generate data
win_pct = np.linspace(0, 1, 1500)
total_faceoffs = np.arange(0, 1500, 1)

# Create a meshgrid
total_faceoffs_grid, win_pct_grid = np.meshgrid(total_faceoffs, win_pct)

# Values of prev to loop over
prev_values = [0, 0.2, 0.4, 0.6, 0.8, 1]

# Create subplots for each value of prev
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

for i, prev in enumerate(prev_values):
    # Evaluate the rating function on the meshgrid
    rating = rating_predictor.cd_face_off_rating(total_faceoffs_grid * win_pct_grid, total_faceoffs_grid * (1 - win_pct_grid), prev)

    # Plot the contour
    ax = axs[i // 3, i % 3]
    contour_plot = ax.contourf(total_faceoffs, win_pct, rating, cmap='Blues')
    ax.set_xlabel('Total Faceoffs')
    ax.set_ylabel('Win Percentage')
    ax.set_title(f'Contour Plot of Faceoff Rating (prev={prev})')
    fig.colorbar(contour_plot, ax=ax, label='Faceoff Rating')

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
