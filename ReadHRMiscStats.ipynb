```python
import pandas as pd
import os


# read the clipboard and skip the first 2 rows

dfBasic = pd.read_clipboard(skiprows=2, header=None, sep=',')

dfBasic
```

```python
# add the custom headers
# Given text string
header_text = "Rk	Player	Age	Tm	Pos	GP	GoalsCreated	GPerGame	APerGame	PtsPerGame	GoalsCreatedPerGame	PIM	SOGPerGameAdjusted	GAdjusted	AAdjusted	PtsAdjusted	GCAdjusted	TotGF	TotPPGF	TotGA	TotPPGA	PlusMin	XGF	XGA	XPlusMin	OPS	DPS	TotPS	ShootOutAtt	ShootoutMade	ShootoutMiss	ShootOutPerc	UID"

# Split the text string into a list of headers
headers = header_text.split()

dfBasic.columns = headers

dfBasic
```

```python
# Add the Season Field and remove goalies

# Add a new column "season" with the value season name

# Prompt the user for the "season" value
season_value = input("Enter the season value (e.g., '22-23'): ")

# Add a new column "season" with the user-inputted value
dfBasic["season"] = season_value

# Remove rows where "Pos" is equal to "G"
dfBasic = dfBasic[dfBasic["Pos"] != "G"]

dfBasic
```

```python
# Drop the columns
# Drop specified columns
columns_to_drop = ["Rk", "Player", "Age", "Tm", "GP", "PIM", "ShootOutAtt", "ShootoutMade", "ShoutoutMiss", "ShootOutPerc"]
dfBasic.drop(columns_to_drop, axis=1, errors='ignore', inplace=True)

dfBasic
```

```python
# Construct the output file path
output_folder = "HockeyRef_Data_Files"
output_file_path = os.path.join(output_folder, f'{season_value}_MiscData.csv')

# Save the DataFrame to a CSV file with the constructed path
dfBasic.to_csv(output_file_path, index=False)
```
